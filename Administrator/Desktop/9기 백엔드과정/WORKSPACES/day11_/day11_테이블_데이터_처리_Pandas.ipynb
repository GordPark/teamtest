# apply

```python
df['성별'].apply(lambda x : '남자' if x ==1 else '여자') # apply()각 행 또는 열에 함수를 적용
# 성별' 열을 기준으로 각 행의 값을 확인하여, 값이 1인 경우 '남자'로 변환하고 그렇지 않은 경우 '여자'로 변환
```

```python
df['성별_str'] = df['성별'].apply(lambda x : '남자' if x ==1 else '여자')
# 변환한 값을 새로운 열 '성별_str'에 할당
```

## groupby

```python
df.groupby('성별') # '성별' 열을 기준으로 그룹화
```

```python
df.groupby('성별').mean() #  '성별' 열의 각 값을 그룹으로 하여 해당 그룹에 속하는 행들의 각 열의 평균값
```

```python
df.value_counts('성별_str') # '성별_str' 열의 각 값의 빈도수를 계산
# df['성별_str'].value_counts()  Series 객체에 대해서 사용
```

```python
df.value_counts('성별_str', normalize=True) # '성별_str' 열의 각 값의 비율을 계산
```

```python
df.groupby('시도코드')['신장(5cm단위)'].mean() 
# '시도코드'를 기준으로 DataFrame을 그룹화한 후, 각 그룹에 속하는 '신장(5cm단위)' 열의 값들의 평균을 계산
```

```python
import logging # 소프트웨어 실행 중에 발생하는 이벤트 디버깅, 모니터링, 오류 분석 등
```

```python
# 기본 로거를 구성하고, 로깅 레벨을 설정
# 레벨 DEBUG / INFO / WARNING / ERROR / CRITICAL
logging.basicConfig(level=logging.ERROR)

# 로깅 메시지 작성
logging.info("This is an info message")
logging.warning("This is a warning message")
logging.error("This is an error message")
```

```python
import logging

# 로거 생성 및 로깅 레벨 설정
logger = logging.getLogger('my_logger')
logger.setLevel(logging.DEBUG)

# 파일 핸들러 생성 및 로거에 추가
file_handler = logging.FileHandler('example.log')
logger.addHandler(file_handler)

# 로깅 메시지 작성
logger.debug("Debug message")
logger.info("Info message")
logger.error("Error message")
```

```python
import logging

# 로거 생성 및 로깅 레벨 설정
logger = logging.getLogger('my_custom_logger') # 로그를 그룹화 'my_custom_logger'라는 이름의 새로운 로거를 생성
logger.setLevel(logging.INFO)

# 핸들러 생성 및 포맷터 설정
handler = logging.StreamHandler()
formatter = logging.Formatter('%(asctime)s - %(name)s - %(levelname)s - %(message)s')
handler.setFormatter(formatter)
logger.addHandler(handler)

# 로깅 메시지 작성
logger.info("This is a custom log message")
```
